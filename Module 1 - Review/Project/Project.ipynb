```python
gamma_ray_data = pd.read_csv("./Homework Data/gamma-ray.csv")
gamma_ray_data[gamma_ray_data["count"] > 0].sort_values(by=["seconds"])
```

```python
# Notes:
# - Gi [count] = Number of gamma rays in a time interval i (rays)
# - ti [seconds] = Length in seconds of time interval i (seconds)
# - lambda = Average rate of gamma rays per second in time interval i (rays / second)

# Calculate the rate of gamma rays per second for each time interval
gamma_ray_data["lambda"] = gamma_ray_data["count"] / gamma_ray_data["seconds"]

# Calculate the average rate of gamma rays per second
average_rate = gamma_ray_data["lambda"].mean()
print("Average rate (MLE)):", average_rate)



```
